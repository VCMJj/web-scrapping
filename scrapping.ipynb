```python
#IMPORT NECESSERY LIBRARY FOR SCRAPPING THE DATA 
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
#PASTE THE URL OF THE WEBSITE YOU WANT TO SCRAPE IN URL VARIABLE
```

```python
url = ("https://quotes.toscrape.com")
```

```python
#USING REQUESTS LIBRARY TO GET THE DATA FROM WEBSITE WEBPAGE 
```

```python
response = requests.get(url) 
```

```python
response = response.content
```

```python
soup = BeautifulSoup(response,'html.parser')
```

```python
#IT WILL FIND THE EXACT PATH THAT WE NEED TO SCRAPE
```

```python
div = soup.find('div')
```

```python
#IT WILL GO THOROUGH THE LIST WHERE OUR ALL THE QUOTE , AUTHOR AND TAGS ARE PRESENT
```

```python
divs = div.find_all('div',class_ = 'quote')
```

```python
#THIS WILL EXTRACT ALL THE TEXT FOR QUOTE BY ITERATE THOROUGH ALL THE QUOTES
```

```python
for div in divs:
    QUOTE_TITLE = div.find('span',class_='text')
    QUOTE_TITLE = QUOTE_TITLE.get_text()
    print(QUOTE_TITLE)
```

```python
#THIS WILL EXTRACT ALL THE AUTHOUR BY ITERATE THOROUGH ALL THE AUTHORS
```

```python
for div in divs:
    Author = div.find('small', class_='author')
    Author = Author.get_text()
    print(Author)
```

```python
#THIS WILL EXTRACT ALL THE TAGS BY ITERATE THOROUGH ALL THE TAGS
```

```python
for div in divs:
    TAGS = div.find('div',class_='tags')
    TAGS = TAGS.get_text()
    print(TAGS)
```

```python
#one loop for executing the code in one go 
```

```python
QUOTE = []
for pages in range (1,5):
    url = (f"https://quotes.toscrape.com/page/{pages}")
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response,'html.parser')
    div = soup.find('div')
    divs = div.find_all('div',class_ = 'quote')
    
    for div in divs:
        QUOTE_TITLE = div.find('span',class_='text')
        QUOTE_TITLE = QUOTE_TITLE.get_text()
        Author = div.find('small', class_='author')
        Author = Author.get_text()
        TAGS = div.find('div',class_='tags')
        TAGS = TAGS.get_text()
        QUOTE.append([QUOTE_TITLE,Author,TAGS])
        
        
    
```

```python
#THIS IS THE FRAME WORK WITH PANDAS 
```

```python
df = pd.DataFrame (QUOTE,columns = ['QUOTE_TITLE','AUTHOR','TAGS'])
```

```python
df['TAGS'] = df['TAGS'].replace('\\n', '|', regex=True)
```
